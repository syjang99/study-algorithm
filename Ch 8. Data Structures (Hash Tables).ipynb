# Hash Tables

## Separate Chaining
- table with each element in linked list form

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None
```

```python
class HashTable:
    def __init__(self):
        self.capacity = 1001
        self.buckets = [None] * self.capacity
        self.count = 0
        
    def hash(self, key):
         return key % self.capacity
        
    def insert(self, key, value):
        index = self.hash(key)
        node = self.buckets[index]
        if node is None:
            self.buckets[index] = Node(key, value)
        else:
            prev = node
            while node is not None:
                prev = node
                node = node.next     #해당 index에 저장된 item 중 마지막 찾기
            prev.next = Node(key, value) #그 다음에 새로 입력할 item 저장
        self.count += 1
        
            
    def retrieve(self, key):
        index = self.hash(key) 
        node = self.buckets[index]
        while node is not None and node.key != key: 
            node = node.next   #해당 key로 저장된 item 나올 때까지 다음 node로 이동
        if node is None: # Not exist
            return None
        else: # Found
            return node.value
        
    def delete(self, key):
        index = self.hash(key)
        node = self.buckets[index]
        prev = None
        while node is not None and node.key != key:
            prev = node
            node = node.next
        if node is None: # Not exist
            return None
        else: # Found
            result = node.value
        if prev is None:   #혼자 있으면 없애버리기
            node = None
        else:
            prev.next = prev.next.next  #중간에 있으면 건너뛰기
        self.count -= 1
        return result
    
    def rehash(self):
        self.capacity *= 2
        new_buckets = [None] * self.capacity
        for item in self_buckets:
            while item is not None:
                new_index = self.hash(item.key)
                new_node = new_buckets[new_index]
                if new_node is None:
                    new_buckets[new_index] = item
                else:
                    prev = new_node
                    while new_node is not None:
                        prev = new_node
                        new_node = new_node.next
                    prev.next = item
        self.buckets = new_buckets
```

## Smallest Missing Positive Integer
- Time complexity: O(N)

```python
def smallest_missing_pos_int(self, list):
    for each item in the list:
        insert into a hash table
    for i = 1, 2, 3, ...:
        if the hash table contains i: keep going
        else: return i
```

```python
def smallest_missing_pos_int(lst):
    max_num = max(lst)
    if max_num <= 0:
        return 1
    
    pos_int = [0]*max_num
    for num in lst:
        if num >0:
            pos_int[num-1] += 1
    for i in range(max_num):
        if pos_int[i] == 0:
            return i+1
```
